```python
#checking and reducing the multicolinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Compute VIF for each feature
vif_data = pd.DataFrame()
vif_data["Feature"] = df.columns
vif_data["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]

print(vif_data)
```

```python
# vif > 10 -> severe multicollinearity should be addressed
# 5 > vif > 10 -> moderate multicolllineratiy
# vif < 5 -> low risk of multicollinerity

#  ways to addressing it
# 1. drop the column with the highest vif
# 2. principal component analysis (PCA)
# 3. regularized regression (Lasso or Ridge)
# 4. domain knowledge(different measures of the same thing, combine them into a single features)
# 5  use model which is more robust to the multicollinearity
```

```python
# im using principal component analysis to reduce teh dimension
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df.drop(columns=['booking_status']))  # Exclude target

pca = PCA(n_components=10)  # Reduce dimensions
X_pca = pca.fit_transform(X_scaled)
```

```python
x = df
y = df['booking_status']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, random_state = 45, test_size = 0.2)
```

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier


classifiers = {
    "Random Forest": RandomForestClassifier(random_state =45),
    # "LogisticRegression": LogisticRegression(random_state = 45, max_iter=500),
    "Gradient Boosting": GradientBoostingClassifier( ),
    "Suppoer Vector classifier": SVC(), 
    "Decission Tree": DecisionTreeClassifier(),
    "KNN": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "Adaboost": AdaBoostClassifier(algorithm='SAMME'),
   
}
```

```python

from sklearn.metrics import f1_score, precision_score, recall_score,accuracy_score


metrics = {
    "Model": [],
    "Accuracy": [],
    "Precision": [],  # Fixed spelling
    "Recall": [],
    "F1 Score": []
}

for model_name, classifier in classifiers.items():
    classifier.fit(x_train, y_train)
    y_predict = classifier.predict(x_test)

    accuracy = accuracy_score(y_test, y_predict)
    recall = recall_score(y_test, y_predict)  # Removed comma
    precision = precision_score(y_test, y_predict)
    f1 = f1_score(y_test, y_predict)

    metrics['Model'].append(model_name)
    metrics['Accuracy'].append(accuracy)
    metrics['Precision'].append(precision)  # Fixed spelling
    metrics['Recall'].append(recall)
    metrics['F1 Score'].append(f1)
```

```python
m_df = pd.DataFrame(metrics)
m_df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV


classifier = RandomForestClassifier()


param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}


grid_search = GridSearchCV(classifier, param_grid, cv=5, scoring='accuracy', n_jobs=-1)


grid_search.fit(x_train, y_train)

best_model = grid_search.best_estimator_


y_pred = best_model.predict(x_test)

print(f"Best Parameters: {grid_search.best_params_}")
```
